```python
class Author:
    def __init__(self, name, birthday, link, quotes=[]):
        self.name = name
        self.birthday = birthday
        self.link = link
        self.quotes = quotes

    def addQuote(self, quote):
        self.quotes.append(quote)

    def toList(self):
        return [self.name, self.birthday, self.link, self.quotes]

    
def tacgiaLink():
    Authors = []
    soup = BeautifulSoup(open('kq.txt', 'r', encoding='utf-8'), "html.parser")
    quotes = soup.find_all('div', class_='quote')
    for quote in quotes:
        name = quote.find('small', class_='author').get_text()
        quote_text = quote.find('span', class_='text').get_text()
        for author in Authors:
            if author.name == name:
                author.addQuote(quote_text.replace('“', '').replace('”', ''))
                break
        else:
            link = "https://quotes.toscrape.com/" + quote.find('a')['href']
            author_page = requests.get(link)
            soup_author = BeautifulSoup(author_page.text, "html.parser")
            birthday = soup_author.find('span', class_='author-born-date').get_text()
            Authors.append(Author(name, birthday, link, [quote_text.replace('“', '').replace('”', '')]))
    return Authors

res = tacgiaLink()
```
